```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
with open("datasets\\brooklyn_sales_map.csv") as f:
    ncols = len(f.readline().split(','))

sales_data = pd.read_csv(filepath_or_buffer="datasets\\brooklyn_sales_map.csv",
                            usecols=range(1,ncols),
                            index_col="zip_code",
                            low_memory=False)

sales_data.head(10)
```

```python
sales_data.dtypes
```

```python
sales_data.isna().sum()
```

```python
have_year_built = sales_data[sales_data.year_built != 0]
have_year_of_sale = have_year_built[have_year_built.year_of_sale != 0]

have_year_of_sale["age"] = have_year_of_sale.year_of_sale - have_year_of_sale.year_built
have_year_of_sale = have_year_of_sale[have_year_of_sale.age > 0]

have_year_of_sale.head(10)
```

```python
mean_age = have_year_of_sale['age'].mean()
total_builts = len(have_year_of_sale.index)
umean_age_count = len(have_year_of_sale[have_year_of_sale.age > mean_age].index)

print(f"Средний возраст построек: {mean_age.round(2)}\n"+
        f"Количество построек: {total_builts}\n"+
        f"Количество построек, возраст которых больше среднего: {umean_age_count}")
```

```python
column_nan_percentge = (have_year_of_sale.isna().sum(axis=0) / len(have_year_of_sale.index) * 100).round(2)
column_nan_percentge
```

```python
sales_data_slice = pd.DataFrame(data=have_year_of_sale,
                                columns=['borough', 'neighborhood', 'building_class_category',
                                            'tax_class', 'building_class', 'gross_sqft',
                                             'total_units', 'age', 'sale_price'])

sales_data_slice.tail(15)
```

```python
row_with_all_na = len(sales_data_slice[sales_data_slice.isna().all(axis=1)].index)

print(f"Количество строк содержащих только NaN / null: {row_with_all_na}")
```

```python
for i in range(0,10):
    sales_data_slice = sales_data_slice.append(pd.Series(dtype='object'), ignore_index=True)

sales_data_slice.tail(15)
```

```python
row_with_all_na = len(sales_data_slice[sales_data_slice.isna().all(axis=1)].index)

print(f"Количество строк содержащих только NaN / null: {row_with_all_na}")
```

```python
sales_data_slice.dropna(how='all', axis=0, inplace=True)

sales_data_slice.tail(15)
```

```python
row_with_all_na = len(sales_data_slice[sales_data_slice.isna().all(axis=1)].index)

print(f"Количество строк содержащих только NaN / null: {row_with_all_na}")
```
